Rozważmy zbiór Boston

```python
import pandas as pd
import numpy as np


data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
#target = raw_df.values[1::2, 2]
feature_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']
```

```python
# print(boston.keys())
# print(boston.data.shape)
# print(boston.feature_names)
bos=pd.DataFrame(data)
bos.head()

bos.columns = feature_names
bos.head()

# print(boston.target[0:5])
bos['PRICE']=target
bos.head()
```

```python
# get the data
boston_X = data
boston_Y = target

# Split the data into training/testing sets
boston_X_train = boston_X[:-50]
boston_X_test = boston_X[-50:]
 
# Split the targets into training/testing sets
boston_y_train = boston_Y[:-50]
boston_y_test = boston_Y[-50:]
```

```python
bos=pd.DataFrame(data[:-50])
bos.columns = feature_names
bos['PRICE']=target[:-50]
bos.head()
```

```python
bos1=pd.DataFrame(boston_X[-50:])
bos1.columns = feature_names
bos1['PRICE']=target[-50:]
bos1.head()
```

```python
model = ols("PRICE ~ I(CRIM)", bos).fit()
# Print the summary
print((model.summary2()))
```

```python
from sklearn import  metrics
metrics.r2_score(bos.PRICE, model.predict(bos))
```
